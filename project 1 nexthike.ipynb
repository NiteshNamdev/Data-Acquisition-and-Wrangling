```python
columns=['season','hr','holiday','weekday','weathersit','temp','hum','windspeed','casual','registered','cnt']
concat_dataset[columns]=concat_dataset[columns].apply(pd.to_numeric)
```

```python
q1= concat_dataset['temp'].quantile(.25)
q3= concat_dataset['temp'].quantile(.75)
```

```python
def find_outliers(df, cols):
    q1 = df[columns].quantile(.25)
    q3 = df[columns].quantile(.75)
    iqr = q3 - q1
    min_r = q1 - 1.5*iqr
    max_r = q3 + 1.5*iqr
    outliers_indices = df.index[(df[columns] < min_r) | (df[columns] > max_r)]
    return outliers_indices
```

```python
find_outliers(concat_dataset,'temp')
```

```python
find_outliers(concat_dataset,'windspeed')
```

```python
find_outliers(concat_dataset,'weathersit')
```

```python
find_outliers(concat_dataset,'cnt')
```

```python
find_outliers(concat_dataset,'hum')
```
